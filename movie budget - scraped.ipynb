```python
from requests import get
from bs4 import BeautifulSoup
import pandas as pd

from pandasql import sqldf
pysqldf = lambda q: sqldf(q, globals())

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style('darkgrid')
```

```python
df_imdb_ratings = pd.read_csv('2016_2019_USmovies.csv')
df_imdb_ratings
```

```python
base_url1 = "https://www.imdb.com/title/"
end_url1 = "/?ref_=adv_li_tt"
```

```python
url = base_url1+df_imdb_ratings.tconst.iloc[6]+end_url1
response = get(url)
print(response)
```

```python
html_soup = BeautifulSoup(response.text, 'html.parser')
movie_info = html_soup
movie_info
```

```python
movie = movie_info.find_all("div", class_ = "txt-block")
movie
```

```python
movie_text = []
for i in range(len(movie)):
    new_string = movie[i].text.replace("\n", "").strip()
    movie_text.append(new_string)
```

```python
for i in range(len(movie_text)):
    if movie_text[i].startswith("Budget") == True:
        budget = movie_text[i].split(":")
        budget = int(budget[1].split(" ")[0].replace("$","").replace(",", ""))
        print(budget)
    elif movie_text[i].startswith("Gross USA") == True:
        gross_usa = movie_text[i].split(":")
        gross_usa = int(gross_usa[1].replace("$","").replace(",", ""))
        print(gross_usa)
    elif movie_text[i].startswith("Cumulative Worldwide Gross") == True:
        gross_world = movie_text[i].split(":")
        gross_world = int(gross_world[1].replace("$","").replace(",", ""))
        print(gross_world)
    else:
        continue
```

```python
df_imdb_ratings["budget_LCY"] = pd.Series([])
df_imdb_ratings["gross_USA_USD"] = pd.Series([])
df_imdb_ratings["cumm_worldwide_gross_USD"] = pd.Series([])
response_dict = {}


for i in range(len(df_imdb_ratings)):
    url = base_url1+df_imdb_ratings.tconst.iloc[i]+end_url1
    response = get(url)
    response_dict[i] = response
    movie_info = BeautifulSoup(response.text, 'html.parser')
    movie = movie_info.find_all("div", class_ = "txt-block")
    movie_text = []
    for j in range(len(movie)):
        new_string = movie[j].text.replace("\n", "").strip()
        movie_text.append(new_string)
    for k in range(len(movie_text)):
        if movie_text[k].startswith("Budget") == True:
            budget = movie_text[k].split(":")
            budget = budget[1].split(" ")[0].replace(",", "")
            df_imdb_ratings["budget_LCY"].iloc[i] = budget
        elif movie_text[k].startswith("Gross USA") == True:
            gross_usa = movie_text[k].split(":")
            gross_usa = int(gross_usa[1].replace("$","").replace("EUR","").replace(",", ""))
            df_imdb_ratings["gross_USA_USD"].iloc[i] = round(gross_usa/1000000,2)
        elif movie_text[k].startswith("Cumulative Worldwide Gross") == True:
            gross_world = movie_text[k].split(":")
            gross_world = int(gross_world[1].replace("$","").replace("EUR","").replace(",", ""))
            df_imdb_ratings["cumm_worldwide_gross_USD"].iloc[i] = round(gross_world/1000000,2)
        else:
            continue
```
